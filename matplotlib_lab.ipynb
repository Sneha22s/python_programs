```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,3,0.01)
y = np.sin(2*np.pi*x)
z = np.cos(2*np.pi*x)
```

```python
fig , ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
ax.set_xlim(0,3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend([r"$sin(2\pi x)$",r"$cos(2\pi x)$"],loc='upper center')
```

```python
x1 = np.arange(0,3,0.01)
y1 = np.sin(2.5*np.pi*x1)
z1 = np.sin(2*np.pi*x1)
fig , ax1 = plt.subplots()
ax1.plot(x1,y1, linewidth=3,color="C1")
ax1.plot(x1,z1,linewidth=3,color="C1")
ax1.set_xlim(0,3)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.fill_between(x1, y1,z1)
```
